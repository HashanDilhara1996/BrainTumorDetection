```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('brain.jpg',0)
```

```python
#Image Histogram
plt.hist(image.flat, bins=100, range = (0,255))
```

```python
ret, th = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
ret, o1 = cv2.threshold(image, 0, 255, cv2.THRESH_TOZERO + cv2.THRESH_OTSU )
```

```python
kernel = np.ones((3,3), np.uint8)
erosion = cv2.erode(o1, kernel, iterations = 1)
```

```python
dilation = cv2.dilate(erosion, kernel, iterations =2)
```

```python
output = [image, o1, erosion, dilation]
    
titles = ['Original', 'OTSU', 'Eroded', 'Dilated']
    
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(output[i],  cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
```

```python
output = [image, dilation]
    
titles = ['Original Image', 'Preprocessed Image']
    
for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(output[i],  cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
```

```python
median = cv2.medianBlur(image,3)
```

```python
output = [image, dilation, median]
    
titles = ['Original Image', 'Preprocessed Image', 'Median Image']
    
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(output[i],  cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
```
